**EV Adoption Trends: Analyze the adoption rate of electric vehicles over the years. How has the demand for EVs changed over time?**

```python
ev_cat['Date'] = pd.to_datetime(ev_cat['Date'], errors='coerce')
ev_cat['Year'] = ev_cat['Date'].dt.year
ev_cat['Month'] = ev_cat['Date'].dt.month
ev_cat['Day'] = ev_cat['Date'].dt.day
ev_cat.head()
```

```python
ev_cat['Date'].info()
```

```python
ev_cat.drop(0,inplace=True)
```

```python
ev_cat['Total EVs']=ev_cat.loc[:,'FOUR WHEELER (INVALID CARRIAGE)':'TWO WHEELER(T)'].sum(axis=1)
ev_cat
```

```python
ev_cat[['Year', 'Total EVs']].sort_values(by='Total EVs',ascending=False)
```

```python
ev_cat[['Year', 'Total EVs']].sort_values(by='Total EVs',ascending=False).head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
x=ev_cat['Year']
y=ev_cat['Total EVs']
plt.plot(x,y, marker='o',color='g')
plt.title('Total EV Adoption Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Number of EVs')
plt.show()
```

**Manufacturer Performance: Compare the performance of different EV manufacturers. Which companies have the highest market share, and what factors contribute to their success?**

**TOTAL SALES BY MAKERS**

```python
ev_sales['total_sales'] = ev_sales.loc[:, '2015':'2024'].sum(axis=1)
performance=ev_sales.groupby('Maker')['total_sales'].sum()
performance
```

```python
performance.sort_values(ascending=False)
```

These companies will have the highest market share cause total_sales for 2014-2024 are very high

```python
performance.sort_values(ascending=False).head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
x=ev_sales['Maker']
y=ev_sales['total_sales']
plt.plot(x,y)
plt.xlabel('Manufacturer')
plt.ylabel('Total Sales')
plt.title('Total Sales by Manufacturer')
plt.show()
```
